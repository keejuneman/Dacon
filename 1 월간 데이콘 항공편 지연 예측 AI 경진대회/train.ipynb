```python
import xgboost as xgb
import optuna
from sklearn.metrics import mean_squared_error



def objective_xgb(trial):
    params = {
        'objective': 'reg:squarederror',
        'eval_metric': 'rmse',
        'booster': 'gbtree',
        'eta': trial.suggest_loguniform('eta', 0.01, 0.1),
        'max_depth': trial.suggest_int('max_depth', 4, 12),
        'subsample': trial.suggest_uniform('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.5, 1),
        'random_state': 42,
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.001, 0.2)
    }

    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    xgb_model = xgb.train(params, dtrain, evals=[(dval, "Validation")], 
                          num_boost_round=1000, early_stopping_rounds=50, verbose_eval=False)
    
    y_pred = xgb_model.predict(dval)
    return mean_squared_error(y_val, y_pred)

study_xgb = optuna.create_study(direction='minimize')
study_xgb.optimize(objective_xgb, n_trials=30)

print("XGBoost - Best trial:")
print(study_xgb.best_trial.params)
print("XGBoost - Best MSE:")
print(study_xgb.best_value)

```

```python
# #XGBoost - Best trial:
# {'eta': 0.041007989248425074, 'max_depth': 4, 'subsample': 0.8977269803624539, 'colsample_bytree': 0.9990951832680529, 'alpha': 0.005678388215612856, 'lambda': 1.7653964437289487, 'learning_rate': 0.043692867967116564}
# XGBoost - Best MSE:
# 0.1020306008162443


best_eta = study_xgb.best_trial.params['eta']
best_max_depth = study_xgb.best_trial.params['max_depth']
best_subsample = study_xgb.best_trial.params['subsample']
best_colsample_bytree = study_xgb.best_trial.params['colsample_bytree']
best_alpha = study_xgb.best_trial.params['alpha']
best_lambda = study_xgb.best_trial.params['lambda']
best_learning_rate = study_xgb.best_trial.params['learning_rate']

xgb_model = xgb.train(params={
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'booster': 'gbtree',
    'eta': best_eta,
    'max_depth': best_max_depth,
    'subsample': best_subsample,
    'colsample_bytree': best_colsample_bytree,
    'alpha': best_alpha,
    'lambda': best_lambda,
    'random_state': 42,
    'learning_rate': best_learning_rate
}, dtrain=xgb.DMatrix(X_train, label=y_train))

y_pred_xgb = xgb_model.predict(xgb.DMatrix(X_val))
mse_xgb = mean_squared_error(y_val, y_pred_xgb)
print("XGBoost - Test MSE:", mse_xgb)
```
